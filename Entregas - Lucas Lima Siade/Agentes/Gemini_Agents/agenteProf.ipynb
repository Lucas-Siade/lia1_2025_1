```python
from dotenv import load_dotenv
import os
import google.generativeai as genai
```

```python
# Carregar vari√°vel de ambiente
load_dotenv()
genai.configure(api_key=os.getenv("GOOGLE_API_KEY"))
```

```python
# Fun√ß√£o para criar agentes simulando comportamentos
def gerar_resposta_gemini(prompt):
    model = genai.GenerativeModel("models/gemini-1.5-flash")
    response = model.generate_content(prompt)
    return response.text.strip()
```

```python
# Agente: Tutor Personalizado
def planejar_estudo(assunto, nivel, tempo_disponivel):
    prompt = f"""
    Voc√™ √© um professor especializado em criar planos de estudo personalizados. 
    Crie um plano de estudo detalhado para o seguinte t√≥pico:
    
    - Assunto: {assunto}
    - N√≠vel de dificuldade: {nivel}
    - Tempo dispon√≠vel por dia: {tempo_disponivel} horas
    
    O plano deve incluir:
    - Objetivos de aprendizado para cada sess√£o de estudo
    - Sugest√µes de recursos (livros, v√≠deos, artigos, etc.)
    - Estrat√©gias de revis√£o
    - Dicas para manter a motiva√ß√£o
    - Distribui√ß√£o de tempo (quantos dias para cada sub-t√≥pico)
    """
    return gerar_resposta_gemini(prompt)
```

```python
# Execu√ß√£o
assunto = "Hist√≥ria do Brasil"
nivel = "b√°sico para alunos do ensino fundamental"
tempo_disponivel = 4

plano_estudo = planejar_estudo(assunto, nivel, tempo_disponivel)

print("üìö Plano de Estudo de Hist√≥ria do Brasil:\n")
print(plano_estudo)
```

```python
# Execu√ß√£o
assunto = "Matem√°tica"
nivel = "avan√ßado para o ITA"
tempo_disponivel = 8

plano_estudo = planejar_estudo(assunto, nivel, tempo_disponivel)

print("üìö Plano de Estudo de Matem√°tica para o ITA:\n")
print(plano_estudo)
```
