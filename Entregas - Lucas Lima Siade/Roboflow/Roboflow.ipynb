#**Extraindo o Dataset**

* [Dataset para Buracos](https://universe.roboflow.com/brad-dwyer/pothole-voxrl/dataset/1)

```python
import zipfile
import os

os.makedirs("/content/Dataset", exist_ok=True)

with zipfile.ZipFile("/content/Pothole.v1-raw.yolov8.zip", 'r') as zip_ref:
    destino = "/content/Dataset/Pothole.v1-raw.yolov8"
    os.makedirs(destino, exist_ok=True)
    zip_ref.extractall(destino)
    print(f"✅ Extraído em {destino}")
```

#**Treinamento do YOLO**

```python
from ultralytics import YOLO
import shutil
import os

model = YOLO('yolov8n.pt')
results = model.train(
    data="/content/Dataset/Pothole.v1-raw.yolov8/data.yaml",
    epochs=50,
    name="modeloBuraco"
)

os.makedirs("/content/drive/MyDrive/ModelosTreinados", exist_ok=True)
shutil.copy("runs/detect/modeloBuraco/weights/best.pt", "/content/drive/MyDrive/ModelosTreinados/bestBuraco.pt")
```

#**Aplicação do Modelo no vídeo**

```python
import cv2, shutil
from ultralytics import YOLO

cap = cv2.VideoCapture("/content/Original.mp4")
out = cv2.VideoWriter(
    "/content/outputBuraco.mp4",
    cv2.VideoWriter_fourcc(*'mp4v'),
    cap.get(cv2.CAP_PROP_FPS),
    (int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)), int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)))
)

model = YOLO("/content/drive/MyDrive/ModelosTreinados/bestBuraco.pt")

while True:
    ret, frame = cap.read()
    if not ret:
        break
    for result in model(frame.copy()):
        for box in result.boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            conf = box.conf[0].item()
            cls = int(box.cls[0].item())
            if conf > 0.5:
                label = f"{model.names[cls]} {conf:.2f}"
                cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
                cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
    out.write(frame)

cap.release()
out.release()
shutil.copy("/content/outputBuraco.mp4", "/content/drive/MyDrive/")
print("✅ Detecção concluída e vídeo salvo no Drive.")
```
